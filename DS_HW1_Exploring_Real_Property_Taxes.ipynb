```python
df.drop('PropertyID', axis=1, inplace=True)
df.drop('Sect', axis=1, inplace=True)
df.drop('PropertyAddress', axis=1, inplace=True)
df.drop('LotSize', axis=1, inplace=True)
df.drop('StateTax', axis=1, inplace=True)
df.drop('ResCode', axis=1, inplace=True)
df.drop('PoliceDistrict', axis=1, inplace=True)
df.drop('CouncilDistrict', axis=1, inplace=True)
df.drop('Location', axis=1, inplace=True)
```

### (b) I think that the AmountDue could be affected by the Neighborhood as well. Some areas may have higher AmountDues, while some others may have lower

```python
colormap1 = {}
for index,name in enumerate(df['Neighborhood'].unique()):
    colormap1[name] = float(index)
df['Neighborhood_ID'] = df['Neighborhood'].apply(lambda x:colormap1[x])
df.plot.scatter(x = 'Neighborhood_ID', y = 'AmountDue')
```

### In the above plot, we can see that the 'AmountDue' is closely distributed with corresponding neighborhoods. This means that most of the properties in same neighborhood will have the same AmountDue as the CityTax would be similar. However, there are some outliers here as well which may constitute properties with pending taxes and larger areas

### I think the third relationship can be between Ward and AmountDue as each property in a ward will be in same city and the CityTax will not vary much.

```python
colormap2 = {}
for index,name in enumerate(df['Ward'].unique()):
    colormap2[name] = float(index)
df['Ward_ID'] = df['Ward'].apply(lambda x:colormap2[x])
df.plot.scatter(x = 'Ward_ID', y = 'AmountDue')
```

### Similar to my comment above, we can see that the Due Amount is nicely distributed among the wards. Yet a few outliers could be seen

### The fourth relationship could be between AmountDue and AsOfDate, as the amount due will be close to a particular date when the tax is generated:

```python
colormap3 = {}
for index,name in enumerate(df['AsOfDate'].unique()):
    colormap3[name] = float(index)
df['Date_ID'] = df['AsOfDate'].apply(lambda x:colormap3[x])
df.plot.scatter(x = 'Date_ID', y = 'AmountDue')
```

### In close relation to my comment above, we can see that the AmountDue is higher and more varied on a particular date, which is most likely the the date when the tax is billed
